```python
import rasterio
import numpy as np
import glob
import os

## Directory where rasters are stored
# raster_dir = "F:\\Biomass tiffs\\Output\\Clippy"

#Updated: 08/05/2025 - to reflect 'final' model
# raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/GEE_Exports"
# raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/log_model"
# raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/LOGO_results"
# raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/LOGO_smoothed"
raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/LOGO_final"


# #All of GA coast:
# raster_dir = "F:/Biomass tiffs/Biomass modeling/Georgia/Georgia model only/all_coast_ga"

raster_files = sorted(glob.glob(os.path.join(raster_dir, "*.tif")))

# Optionally extract years from filenames
# years = [int(os.path.basename(f)[4:8]) for f in raster_files] #For non-clipped output
years = [int(os.path.basename(f).split('_')[-1].split('.')[0]) for f in raster_files]

##ADDED
years = np.array(years)

# Load stack: shape (time, rows, cols)
stack = []
for file in raster_files:
    with rasterio.open(file) as src:
        stack.append(src.read(1).astype(np.float32))  # Ensure float for NaNs
        profile = src.profile  # Save profile from one file to write output later

stack = np.stack(stack)  # (time, rows, cols)

# Mask no-data values (assuming 0 or some other value is nodata)
stack[stack == profile['nodata']] = np.nan  # Or pick your own nodata value



    ##OLD
# # Read all rasters into a 3D array (time, rows, cols)
# stack = []
# for file in raster_files:
#     with rasterio.open(file) as src:
#         stack.append(src.read(1))  # Read first band

# stack = np.stack(stack)  # Shape: (years, rows, cols)
```

```python
'''##Non-parametric technique with considerations for p-values:
import os
import glob
import numpy as np
import rasterio
import pymannkendall as pmk
from statsmodels.stats.multitest import multipletests

rows, cols = stack.shape[1:]

# Initialize empty arrays for the Theil-Sen slope and raw p-values
slope_array = np.full((rows, cols), np.nan, dtype=np.float32)
pval_array = np.full((rows, cols), np.nan, dtype=np.float32)

print("Calculating Yue-Pilon pre-whitened Mann-Kendall and Theil-Sen slopes per pixel...")

# 1. Calculate non-parametric slope + p-value per pixel
for i in range(rows):
    # Optional progress tracker since this takes longer than OLS
    if i % 100 == 0: 
        print(f"Processing row {i} of {rows}...")
        
    for j in range(cols):
        y = stack[:, i, j]

        # Remove NaNs BEFORE running Mann-Kendall
        y_clean = y[np.isfinite(y)]

        if len(y_clean) >= 20 and np.std(y_clean) > 0:
            try:
                res = pmk.yue_wang_modification_test(y_clean)

                slope_array[i, j] = res.slope
                pval_array[i, j] = res.p

            except Exception as e:
                print(f"Error at pixel ({i},{j}): {e}")

# 2. Save the Master Theil-Sen Slope Raster
slope_out = "F:/Biomass tiffs/Biomass modeling/Georgia/All_trend/LOGO_1984to2023_Thiel_Sen.tif"
profile.update(dtype=rasterio.float32, count=1, nodata=np.nan)
with rasterio.open(slope_out, 'w', **profile) as dst:
    dst.write(slope_array, 1)
print(f"Master Theil-Sen Slope map saved: {slope_out}")

# ==========================================
# 3. APPLY BENJAMINI-HOCHBERG FDR CORRECTION
# ==========================================
print("Applying Benjamini-Hochberg False Discovery Rate adjustment...")

# Isolate valid, calculated p-values to feed into the multiple-test corrector
valid_p_mask = ~np.isnan(pval_array)
flat_p_values = pval_array[valid_p_mask]

# Initialize a blank boolean array for the final significance mask
sig_mask = np.zeros_like(pval_array, dtype=bool)

if len(flat_p_values) > 0:
    # reject: boolean array showing True where trends survive FDR correction at alpha=0.05
    reject, _, _, _ = multipletests(flat_p_values, alpha=0.05, method='fdr_bh')
    
    # Map the 1D results back onto your 2D spatial grid
    sig_mask[valid_p_mask] = reject

# ==========================================
# 4. SAVE THE SIGNIFICANT SLOPES RASTER
# ==========================================
sig_slopes = np.where(sig_mask, slope_array, np.nan)
sig_out = "F:/Biomass tiffs/Biomass modeling/Georgia/All_trend/LOGO_sig_1984to2023_Thiel_Sen.tif"
with rasterio.open(sig_out, 'w', **profile) as dst:
    dst.write(sig_slopes, 1)

print(f"FDR-Significant slope map saved: {sig_out}")'''
```

```python
##Hotspot pixel count

##FLATS
base_layer1 = "F:/Flats imagery/Trend_nocutoffareas/flood84_23.tif"

##LOGO
# base_layer = "F:/Biomass tiffs/Biomass modeling/Georgia/All_trend/LOGO_sig_1984to2023_final.tif"
base_layer2 = "F:/Biomass tiffs/Biomass modeling/Georgia/All_trend/LOGO_sig_1984to2023_Thiel_Sen.tif"

base_shape = "F:/Wetlands shapefiles/GA_shapefile_wetlands/GA_estuaries.shp"
gdf = gpd.read_file(base_shape)

import rasterio
from rasterio.mask import mask
import geopandas as gpd

# 2. Ensure the shapefile CRS matches your rasters
# (It is critical that both rasters and the shapefile are in the same CRS)
with rasterio.open(base_layer1) as src:
    gdf = gdf.to_crs(src.crs)
    shapes = [feature for feature in gdf.geometry]

# 3. Create a function to clip/mask both to the shapefile
def clip_raster_to_shape(raster_path, shapes):
    with rasterio.open(raster_path) as src:
        # crop=True cuts the raster to the shapefile's bounding box
        out_image, out_transform = mask(src, shapes, crop=True)
        
        # Select ONLY the first band [0, :, :]
        # This turns the (bands, height, width) array into a (height, width) array
        return out_image[0, :, :], out_transform

# Clip both
data1, transform1 = clip_raster_to_shape(base_layer1, shapes)
data2, transform2 = clip_raster_to_shape(base_layer2, shapes)

# 4. Now both data1 and data2 are 2D arrays (height, width)
if data1.shape != data2.shape:
    print(f"Still a shape mismatch! Raster1: {data1.shape}, Raster2: {data2.shape}")
else:
    # Since they are now 2D, you don't need the [0] index anymore
    mask_condition = (data1 > 0.005) & (data2 < 0) ##0.005 chosen because 40*0.005 is 0.2, the threshold for FLATS+. Absent sig
    print(f"Pixel count: {np.sum(mask_condition)}")
```
